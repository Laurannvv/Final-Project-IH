```python
#Import tools
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml
import re
import numpy as np
```

```python
#Define function

def get_laws_p1_only(url):
    html = requests.get(url).content
    soup = BeautifulSoup(html, "lxml")
    urls = list(soup.find_all('h2'))
    txt1=[]
    titles=[]
    txt2=[]
    for url in urls:
        txt1.append(re.findall('https://?[\w/\-?=%.]+\.[\w/\-?=%.:\d]+', str(url)))
    for title in urls:
        titles.append(re.findall('>\w+.+<',str(title)))
    for line in titles:
        for el in line:
            txt2.append(re.sub('[><]|/a','',el))
    document_list= pd.DataFrame({
        'Titles': txt2[1:],
        'URL': txt1[1:]})
    return document_list
    
    
```

```python
#Extraction code

txt1=[]
titles=[]
txt_title=[]
urls_extracted=[]
big_soup =[]

pages = np.arange(2,5)

#Extracting data page 1

get_laws_p1('https://eur-lex.europa.eu/search.html?SUBDOM_INIT=ALL_ALL&DTS_DOM=ALL&type=advanced&DTS_SUBDOM=ALL_ALL&excConsLeg=true&qid=1607349508496&locale=en#')

#Extracting data for every other pages

for page in pages:
    page_extract = requests.get("""https://eur-lex.europa.eu/search.html?SUBDOM_INIT=ALL_ALL&DTS_DOM=ALL&type=advanced&DTS_SUBDOM=ALL_ALL&excConsLeg=true&qid=1607349508496&locale=en&page="""+str(page))
    soup = BeautifulSoup(page_extract.text, 'html.parser')
    urls = soup.find_all('h2')
    
    for url in urls:
        txt1.append(re.findall('https://?[\w/\-?=%.]+\.[\w/\-?=%.:\d]+', str(url)))
        
    
    for title in urls:
        titles.append(re.findall('>\w+.+<',str(title)))
        
for line in txt1:
    for url in line:
        urls_extracted.append(url)
        
for line in titles:
    for el in line:
        txt_title.append(re.sub('[><]|/a|Search criteria','',el))
        if '' in txt_title:
            txt_title.remove('')

document_list= pd.DataFrame({
    'Titles': txt_title,
    'URL': urls_extracted})

#Cleaning DataFrame

document_list.drop(document_list.loc[(document_list.Titles == 'Search criteria') | (document_list.URL == """[]""")].
                   index, inplace = True)

document_list.reset_index(drop= True, inplace = True)

document_list['Titles'] = document_list[['Titles']].apply(lambda x : x.str.lstrip())

document_list

```
