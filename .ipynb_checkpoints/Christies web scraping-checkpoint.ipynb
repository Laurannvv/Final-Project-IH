```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def get_christies_sale(saleid,year,month):
    url = 'https://www.christies.com/lotfinder/print_sale.aspx?saleid='+str(saleid)+'&lid=1'
    html = requests.get(url).content
    #html[0:600]
    soup = BeautifulSoup(html, "lxml")

    #soup
    text1 = [element.text for element in soup.find_all('tr')]
    #text1
    nmbrtxt = text1[0].split('for Sale')[1].strip()
    #nmbrtxt
    text1 = text1[1:]
    text2 = [line.replace('\n\n\n\n\n','') for line in text1]
    text2
    text3 = [line.split('\n')+[line] for line in text2]
    text4 = []
    for line in text3:
        row=[line[0],'','','','','','','','']
        row[0] = line[0] #Lot
        if (len(line) == 9) and (line[4]!=''): # regular fine arts
            row[1] = line[1] #Artist
            row[2] = line[2] #Title/Description
            row[3] = line[3] #Medium
            row[4] = line[4] #Size
            row[5] = line[5] #X
            row[6] = line[6] #Estimate
            row[7] = line[7] #Realized
            row[8] = line[8] #Full text of the line

        elif (len(line) == 8) and (line[4]==''): # prints?
            row[1] = line[1] #Artist
            row[2] = line[2] #Title/Description
            row[3] = ''      #Medium
            row[4] = line[3] #Size
            row[5] = line[4] #X
            row[6] = line[5] #Estimate
            row[7] = line[6] #Realized
            row[8] = line[7] #Full text of the line

        elif (len(line) == 7) and (line[3]==''): # drawings?
            row[1] = line[1] #Artist
            row[2] = line[2] #Title/Description
            row[3] = '' #Medium
            row[4] = '' #Size
            row[5] = line[3] #X
            row[6] = line[4] #Estimate
            row[7] = line[5] #Realized
            row[8] = line[6] #Full text of the line

        elif (len(line) == 10): # prints?
            row[1] = line[1] #Artist
            row[2] = line[2] #Title/Description
            row[3] = line[3]+' '+line[4] #Medium
            row[4] = line[5] #Size
            row[5] = line[6] #X
            row[6] = line[7] #Estimate
            row[7] = line[8] #Realized
            row[8] = line[9] #Full text of the line
            
        else: # watches, miscellaneous
            row[2] = line[1] #no artist, only description
            if (line[2].startswith('BY ')) or (line[2].startswith('By ')) or (line[2].startswith('by ')):
                row[1] = line[2]
            if (line[2].startswith('--')) or  ('in.' in line[2]) or ('cm.' in line[2]):
                row[4] = line[2]
            if (line[3].startswith('--')) or  ('in.' in line[3]) or ('cm.' in line[3]):
                row[4] = line[3]
            if (line[4].startswith('--')) or  ('in.' in line[4]) or ('cm.' in line[4]):
                row[4] = line[4]
            if (line[5].startswith('--')) or  ('in.' in line[5]) or ('cm.' in line[5]):
                row[4] = line[5]
            row[6] = line[-3] #Estimate
            row[7] = line[-2] #Realized
            row[8] = line[-1] #Full text of the line 
        text4.append(row)
    colnames = ['Lot', 'Artist', 'Title_descr', 'Medium', 'Size', 'X', 'Estimate', 'Realized','Full_text']
    df = pd.DataFrame(text4, columns=colnames)
    df['Auction_house'] = 'Christies'
    df['Sale_id'] = nmbrtxt
    df['Sale_id_internal'] = saleid
    df['Year'] = year
    df['Month'] = month
    df = df[['Auction_house','Sale_id', 'Sale_id_internal', 'Year', 'Month', 'Lot', 
             'Artist', 'Title_descr', 'Medium', 'Size', 'Estimate', 'Realized', 'Full_text']]
    return df
```
